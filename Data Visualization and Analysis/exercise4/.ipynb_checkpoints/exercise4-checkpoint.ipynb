# Data visualization

*Using `matplotlib` to visuialize historical Fortune500 data*

This assignment is individual and you agree to submit your own work.


```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

from pandas import Series, DataFrame

np.set_printoptions(precision=2, suppress=True, linewidth=120)

pd.options.display.max_rows = 20

plt.rc("figure", figsize=(8, 4))
```

## Data preparation

Read the data from the provided *fortune500.csv* into a `DataFrame`.

Drop "(in millions)" from the names of the columns so that dot-notation can be used to access them.

Set data type of the *Company* column to a string and data type of the "Profit (in millions)" to a floating-point number.

```python
df = pd.read_csv("fortune500.csv")

df.rename({"Revenue (in millions)": "Revenue", "Profit (in millions)": "Profit"},
          axis="columns",
          inplace=True)

df.Company = df.Company.astype(pd.StringDtype())
df.Profit = pd.to_numeric(df.Profit, errors="coerce")
```

## Task 1

Plot the historical revenue of three companies, *General Dynamics*, *General Electric*, and *General Motors*.

The figure must have 4 subplots, 3 companies individually and 1 plot with all three companies.

`x`-axis on each plot must be the year and `y`-axis must be the revenue of the company that year. It's easier to achieve if `Year` is an index of the series with a company revenue.

Your figure should look similar (though not necessarily identical) to the following:

![Task 1](task1.png)

```python
# Add your implementation here.
companies = ["General Dynamics", "General Electric", "General Motors"]

fig = plt.figure(figsize=(18, 8))
for i, company in enumerate(companies):
    company_df = df[df['Company'] == company]
    
    if i == 0:
        color = "red"
    elif i == 1:
        color = "green"
    elif i == 2:
        color = "blue"
        
    ax = fig.add_subplot(2, 3, i+1)
    ax.plot(company_df['Year'], company_df["Revenue"], color=color)
    ax.set_ylabel("Revenue (millions of dollars)")
    ax.set_title(company)

ax_comp = fig.add_subplot(2, 1, 2)
for i, company in enumerate(companies):
    company_df = df[df['Company'] == company]
    
    if i == 0:
        color = "red"
    elif i == 1:
        color = "green"
    elif i == 2:
        color = "blue"
        
    ax_comp.plot(company_df['Year'], company_df["Revenue"], color=color)
    ax_comp.set_ylabel("Revenue (millions of dollars)")
    ax_comp.set_title("Comparison")


fig.subplots_adjust(wspace=0.5, hspace=0.3)
```

## Task 2

Plot the total profits of *Hershey Foods* by decade as a bar chart.

A decade is ten years that share the first 3 digits, so the figure must include 6 decades: 50s, 60s, 70s, 80s, 90s, and 00s.

Your figure should look similar (though not necessarily identical) to the following:

![Task 2](task2.png)

```python
# Add your implementation here.
df['decade'] = (df['Year'] // 10) * 10

decade_prof = df[df['Company'] == 'Hershey Foods'].groupby('decade')['Profit'].sum()
decade_labels = ['50s', '60s', '70s', '80s', '90s', '00s']

fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(decade_labels, decade_prof, label="Profit")

ax.set_title('Hershey Foods')
ax.set_xlabel('Decade')
ax.set_ylabel('Profit (mxillions of dollars)')
ax.legend()
```

## Task 3

Plot the revenue and rank of *Microsoft* as a scatterplot.

Revenue must be the `x`-axis, rank must be the inverted `y`-axis.

Your figure should look similar (though not necessarily identical) to the following:

![Task 3](task3.png)


```python
# Add your implementation here.
fig, ax = plt.subplots(figsize=(9, 5))

ax.scatter(df[df['Company'] == 'Microsoft']['Revenue'], df[df['Company'] == 'Microsoft']['Rank'])
ax.invert_yaxis()

ax.set_xlabel('Revenue (millions of Dollars)')
ax.set_ylabel('Rank')
ax.set_title('Microsoft')
```

## Task 4

Plot IBM's (*Intl. Business Machines*) annual rank between 1965 and 2005, annotating the following dates:

- 1970: "Relational DB"
- 1981: "IBM PC"
- 1997: "Deep Blue"

The `y`-axis should be inverted and limited to `[0..15]` to make changes in ranking more visible.

Your figure should look similar (though not necessarily identical) to the following:

![Task 4](task4.png)

```python
# Add your implementation here.
ibm_df = df[(df["Year"] >= 1965) & (df["Year"] <= 2005) & (df["Company"] == "Intl. Business Machines")]
plt.gca().invert_yaxis()
plt.ylim(15,0)
plt.xlim(1965, 2005)

plt.title("IBM")
plt.ylabel("Rank")
plt.xlabel("Year")

years = ibm_df.Year
ranks = ibm_df.Rank

annotations = [(1970, "Relational DB"), (1981, "IBM PC"), (1997, "Deep Blue")]

# Add annotations to the plot with arrows
for xval, lbl in annotations:
    if xval in years.values:
        yval = ranks[years == xval].values[0]
        plt.annotate(lbl, xy=(xval, yval + 1), xytext=(xval, yval + 4),
                     arrowprops={"facecolor": "black", "width": 2, "headwidth": 4, "headlength": 4},
                     horizontalalignment="left", verticalalignment="top")

plt.plot(years, ranks)
plt.show()
```

## Task 5

Plot Apple's (*Apple Computers*) annual profit and mark Steve Jobs's absence (1985-1997) as a red zone (rectangle).

Your figure should look similar (though not necessarily identical) to the following:

![Task 5](task5.png)

```python
# Add your implementation here.
apple_df = df[df["Company"] == "Apple Computer"]

plt.title("Apple")
plt.ylabel("Profit (millions of dollars")

years = apple_df.Year
profit = apple_df.Profit

y_min, y_max = plt.ylim(1000,-1200)
rect = matplotlib.patches.Rectangle((1985, y_min), 12, y_max-y_min, color="red", alpha=0.5)
plt.gca().add_patch(rect)
plt.gca().invert_yaxis()
plt.annotate("Jobs gone", xy=(1990, -500))

plt.plot(years, profit)
plt.show()
```
