```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("library.csv")
```

```python
df.dtypes
```

```python
#inpace is invasive
df.drop("url", axis=1, inplace=True)
```

```python
#thresh value is the value of non NA values that a row has to not be dropped
df.dropna(thresh=6)
```

```python
df.pages = np.around(df.pages.fillna(df.pages.mean()),0)
```

```python
df.price = np.around(df.price.fillna(df.price.mean()),0)
```

```python
df.year.fillna(method ="ffill")
```
