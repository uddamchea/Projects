## Seaborn

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

#specify option
plt.rc("figure", figsize=(8,4))
```

```python
sample = sns.load_dataset("iris")
```

```python
sample.info()
```

```python
sample.head(3)
```

```python
fig, axis = plt.subplots(2, 2)
```

```python
axis
```

```python
axis[0,0].plot(sample.sepal_length)
axis[0,1].plot(sample.petal_length)
axis[1,0].plot(sample.sepal_width)
axis[1,1].plot(sample.petal_width)
fig
```

## Pandas

```python
s = pd.Series(np.random.standard_normal(10).cumsum(),
              index = np.arange(0, 100, 10))
s
```

```python
s.hist()
```

```python
s.plot.hist()
```

```python
s.plot(kind = "hist")
```

```python
s.plot(kind = "bar", title="Title of Chart")
```

```python
s.plot(kind = "bar", title="Title of Chart", grid=True, rot=0)
```

```python
df=pd.DataFrame(np.random.standard_normal((10,4)).cumsum(0),
                columns = ["A", "B", "C", "D"],
                index = np.arange(0,100,10))
```

```python
plt.style.available
```

```python
df.plot()
```

```python
plt.style.use("fivethirtyeight")
df.plot()
```

```python
df.plot(subplots=True)
```

```python
df.plot.bar(subplots=True)
```
