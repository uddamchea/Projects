```python
import pandas as pd
import numpy as np
import matplotlib as plt
from matplotlib import pyplot
```

```python
#defined file path
url="https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
#everything in file
df=pd.read_csv(url)
#Without header
dfwh=pd.read_csv(url,header= None )
#add headers in list
headers=["symboling","normalized-losses","make","fuel-type","aspiration","num-of-doors","body-style","drive-wheels",
         "engine-location","wheel-base","length","width","height","curb-weight","engine-type","num-of-cylinders",
         "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower","peak-rpm","city-mpg","highway-mpg","price"]
#assign list of header as data frame column names
dfwh.columns=headers
```

```python
#prints dataframe attributes data types in series
print(dfwh.dtypes)
```

```python
#Statistical analysis of data distribution
print(dfwh.describe())
```

```python
#It describes top to and bottom 30 rows each for sample
print(dfwh.info)
print(dfwh)
```

```python
#Data cleaning is done on columns
#You cana ccess each column by referening its index or name
#iloc takes two arguments i.e. # of rows or row index, column index or name
print(dfwh.iloc[:,0])
#you can directly refere to column and read all data
print(dfwh['symboling'])
```

```python
#can perform opertaion on column as well
dfwh['symboling']=dfwh['symboling']+1
print(dfwh['symboling'])
```

```python
#****************************************************************************
#Data cleaninig examples
#****************************************************************************
print(dfwh[['highway-mpg','price']])
#drop missing values attribute
dfwh.dropna(subset=["price"],axis=0,inplace=True)
print(dfwh[['highway-mpg','price']])
```

```python
#drop missing value on certain value
ndf=dfwh[dfwh["price"]=="?"]
dfwh.drop(ndf.index,axis=0,inplace=True)
print(dfwh[['highway-mpg','price']])
print(ndf)
```

```python
#replace missing values
dfwh["normalized-losses"].replace("?","0",inplace=True)
dfwh[["normalized-losses"]] = dfwh[["normalized-losses"]].apply(pd.to_numeric)
meanNL=dfwh["normalized-losses"].mean()
dfwh["normalized-losses"].replace(0,meanNL,inplace=True)
print(meanNL)
print(dfwh['normalized-losses'])
```

```python
#Data format conversion
print(dfwh["city-mpg"])
dfwh["city-mpg"]=235/dfwh["city-mpg"]
dfwh.rename(columns={'city-mpg': 'city-L/100KM'}, inplace=True)
print(dfwh["city-L/100KM"])
```

```python
#Data normalziation
#simple feature scaling
dfsfc=dfwh
dfmm=dfwh
dfz=dfwh
print(dfwh["length"])
dfsfc["length"]=dfsfc["length"]/dfsfc["length"].max()
print(dfsfc["length"])
dfmm["length"]= (dfmm["length"]-dfmm["length"].min())/(dfmm["length"].max()-dfmm["length"].min())
print(dfmm["length"])
dfz["length"]=(dfz["length"]-dfz["length"].mean())/dfz["length"].std()
print(dfz["length"])
```

```python
#Data Binning
#We want to divide the range of the data into 3 equally sized bins, but the function cut 
#requires bins
dfwh["price"]=dfwh["price"].astype('int')
binwidth=int((max(dfwh["price"])-min(dfwh["price"]))/3)
print(binwidth)
#we build the bin array, from min value to max value, using bin width calculated above.
bins=range(min(dfwh["price"]),max(dfwh["price"]),binwidth)
print(bins)
#we create labels for each bin: “Low”, “Medium” and “High”.
price_groups=np.array(['low', 'medium', 'high'])
#we create a column similar to the one with the group price of a car.
dfwh["price_group"]=pd.cut(dfwh["price"],bins=bins, labels=price_groups)
print(dfwh[['price_group','price']])
# draw historgram of attribute "horsepower" with bins = 3
plt.pyplot.hist(dfwh["price"], bins = 3)
# set x/y labels and plot title
plt.pyplot.xlabel("Price")
plt.pyplot.ylabel("count")
plt.pyplot.title("Price bins")
```

```python
#Categorical to numbers
dummy_variable_1 = pd.get_dummies(dfwh["fuel-type"])
dummy_variable_1.head()
dummy_variable_1.rename(columns={'fuel-type-diesel':'gas','fuel-type-diesel':'diesel'}, inplace=True)
dummy_variable_1.head()
# merge data frame "df" and "dummy_variable_1" 
dfwh = pd.concat([dfwh, dummy_variable_1], axis=1)
# drop original column "fuel-type" from "df"
dfwh.drop("fuel-type", axis = 1, inplace=True)
dfwh.head()
print(dfwh[['make','diesel','gas']])
```
