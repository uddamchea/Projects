```python
#Removing attributes with little to no values
new_df = df.drop(['birthplace_gold', 'date_of_birth_gold', 'race_ethnicity_gold', 'religion_gold', 'sexual_orientation_gold', 'year_of_award_gold'], axis=1)
print(new_df.shape)
```

```python
#Filling in missing values for all attributes
##Using 'ffill' for _last_judgment_at and 'bfill' for religion
new_df[['_last_judgment_at']].fillna(method='bfill', inplace = True)
new_df[['religion']].fillna(method='ffill', inplace = True)
new_df.head(441)
```

```python
#Finding duplicate rows
dup_df=new_df[new_df.duplicated()]
print('There are ', dup_df.shape, 'duplicate rows')
```

```python
#Bucketing year_of_award 
year_award_label = ['New', 'Old', 'Too Old']
category = [1850,1900,2000,2020]
new_df['year_of_award_label'] = pd.cut(new_df['year_of_award'], labels=year_award_label, bins=category, include_lowest = False)
print(new_df[['year_of_award','year_of_award_label']])
```

```python
#Bucketing birthplace_confidence
birthplace_label = ['Unsure', 'Sure']
category = [0.,.99,1.]
new_df['birthplace:confidence_label'] = pd.cut(new_df['birthplace:confidence'], labels=birthplace_label, bins=category, include_lowest = False)

print(new_df[['birthplace:confidence','birthplace:confidence_label']])
```
